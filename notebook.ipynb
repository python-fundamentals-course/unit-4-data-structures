## 1. Lists — ordered, changeable collections

You've already used lists in Units 2 and 3. A **list** is an ordered collection that can grow, shrink, and be modified after it's created. Items are accessed by their position (**index**), starting from `0`.


````python
fruits = ["apple", "banana", "cherry"]

print(fruits[0])        # first item
print(fruits[-1])       # last item

fruits.append("durian")
print("After append:", fruits)

fruits[1] = "blueberry"
print("After change:", fruits)
````

## 2. Tuples — ordered, unchangeable collections

A **tuple** looks like a list but uses `( )` instead of `[ ]`, and once created it **cannot be changed** (it's *immutable*). Tuples are useful for fixed groups of values — like coordinates, RGB colours, or a record that should never be accidentally modified.


````python
coordinates = (1.3521, 103.8198)  # latitude, longitude for Singapore
print(coordinates)
print("Latitude:", coordinates[0])

# This would raise an error if uncommented — tuples can't be changed:
# coordinates[0] = 0.0

# Tuples are often used to return multiple values from a function (Unit 3)
# and to 'unpack' several values at once:
latitude, longitude = coordinates
print(f"lat={latitude}, lon={longitude}")
````

## 3. Sets — unique, unordered collections

A **set** stores items with **no duplicates** and **no guaranteed order**. Sets are ideal for membership checks ("is this value in here?") and for removing duplicates from a list.


````python
tags = {"python", "data", "beginner", "python"}  # duplicate is dropped
print(tags)

print("python" in tags)   # fast membership check

# Removing duplicates from a list using set()
registrations = ["alice", "bob", "alice", "carol", "bob", "alice"]
unique_attendees = set(registrations)
print(f"{len(registrations)} registrations, {len(unique_attendees)} unique attendees")
````

## 4. Dictionaries — key/value pairs

A **dict** stores data as `key: value` pairs. Instead of looking items up by position (like a list), you look them up by a meaningful **key** — making dicts the natural way to represent a single record (e.g. one customer, one product, one transaction).


````python
student = {
    "name": "Mei Lin",
    "age": 27,
    "course": "Python Fundamentals",
    "completed_units": 4,
}

print(student["name"])
print(student["completed_units"])

# Adding or updating a key is just an assignment
student["completed_units"] = 5
student["email"] = "mei.lin@example.com"
print(student)

# .get() avoids an error if the key doesn't exist
print(student.get("phone", "no phone on file"))
````

## 5. Choosing the right data structure

| Structure | Ordered? | Changeable? | Duplicates? | Looked up by | Good for |
|---|---|---|---|---|---|
| **list** | yes | yes | yes | position (index) | a sequence of similar items |
| **tuple** | yes | no | yes | position (index) | a fixed group of values |
| **set** | no | yes | no | membership (`in`) | unique items, fast lookups |
| **dict** | yes (insertion order) | yes | keys must be unique | key | a single record's named fields |

**Rule of thumb:** if each item has *named fields* (like `name`, `price`, `date`), reach for a **dict**. If you have a *collection* of similar items, reach for a **list**.


## 6. The list-of-dicts pattern

Combine the two: a **list of dicts** represents a table of data — each dict is one row, and each key is a column. This is the pattern you'll use constantly when working with real data, and it's exactly what `pandas` (which you'll preview in Unit 6) builds on.


````python
sales = [
    {"date": "2026-01-05", "product": "Laptop", "amount": 1500},
    {"date": "2026-01-06", "product": "Mouse", "amount": 25},
    {"date": "2026-01-06", "product": "Laptop", "amount": 1500},
    {"date": "2026-01-07", "product": "Monitor", "amount": 300},
    {"date": "2026-01-08", "product": "Mouse", "amount": 25},
]

for row in sales:
    date = row["date"]
    product = row["product"]
    amount = row["amount"]
    print(f"{date}: {product} - ${amount}")
````

## 7. List comprehensions

In Unit 2 you wrote loops to **filter** and **transform** lists. A **list comprehension** is a compact way to write the same thing in a single line. It reads almost like the plain-English description of what you want.

```python
new_list = [expression for item in iterable if condition]
```


````python
amounts = [25, 1200, 80, 450, 3000]

# Loop version (from Unit 2)
large_loop = []
for amount in amounts:
    if amount > 100:
        large_loop.append(amount)

# Comprehension version — same result, one line
large_comp = [amount for amount in amounts if amount > 100]

print("Loop:        ", large_loop)
print("Comprehension:", large_comp)

# Comprehensions can transform too
doubled = [amount * 2 for amount in amounts]
print("Doubled:      ", doubled)

# And work directly with a list of dicts
laptop_sales = [row["amount"] for row in sales if row["product"] == "Laptop"]
print("Laptop sales: ", laptop_sales)
````

## 8. Dict operations for processing data

A few dict patterns come up again and again when summarising a list-of-dicts:

- Build up **totals per category** using a dict as a running tally
- Use `.keys()`, `.values()`, and `.items()` to loop over a dict's contents
- Use `dict.setdefault()` or check `if key not in dict` to handle a key you haven't seen yet


````python
sales = [
    {"date": "2026-01-05", "product": "Laptop", "amount": 1500},
    {"date": "2026-01-06", "product": "Mouse", "amount": 25},
    {"date": "2026-01-06", "product": "Laptop", "amount": 1500},
    {"date": "2026-01-07", "product": "Monitor", "amount": 300},
    {"date": "2026-01-08", "product": "Mouse", "amount": 25},
]

# Total sales per product, using a dict as a running tally
totals_by_product = {}
for row in sales:
    product = row["product"]
    amount = row["amount"]
    if product not in totals_by_product:
        totals_by_product[product] = 0
    totals_by_product[product] += amount

# Loop over the resulting dict with .items()
for product, total in totals_by_product.items():
    print(f"{product}: ${total}")

print("\nAll products:", list(totals_by_product.keys()))
print("All totals:  ", list(totals_by_product.values()))
````
